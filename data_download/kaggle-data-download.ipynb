```plaintext
DATA_SPLIT = "train"
PLANET_IDS = ["1010375142", "1024292144", "1029552010", "1031303815", "1042982756", "1047977648", "1048114509", "104891231", "1049092982"]
OUTPUT_FILE = "data.zip"
```

```plaintext
import zipfile
import os
from pathlib import Path
from IPython.display import FileLink
```

```plaintext
folders_to_zip = [Path("/kaggle/input/ariel-data-challenge-2025") / DATA_SPLIT / planet_id for planet_id in PLANET_IDS]
```

```plaintext
def zip_multiple_folders(output_zip_name, folders_to_zip):
    """
    Zips multiple specified folders into a single archive.

    Args:
        output_zip_name (str): The name of the output .zip file.
        folders_to_zip (list): A list of paths to the folders to be zipped.
    """
    try:
        with zipfile.ZipFile(output_zip_name, 'w', zipfile.ZIP_DEFLATED) as zipf:
            for folder_path in folders_to_zip:
                if not os.path.exists(folder_path) or not os.path.isdir(folder_path):
                    print(f"Warning: Folder '{folder_path}' not found or is not a directory. Skipping.")
                    continue

                for root, _, files in os.walk(folder_path):
                    for file in files:
                        file_full_path = os.path.join(root, file)
                        # Create an archive name that maintains the folder structure relative to the parent of the input folder
                        relative_path_in_zip = os.path.relpath(file_full_path, os.path.dirname(folder_path))
                        zipf.write(file_full_path, relative_path_in_zip)
        print(f"Successfully created '{output_zip_name}' containing the specified folders.")
    except Exception as e:
        print(f"An error occurred: {e}")
```

```plaintext
zip_multiple_folders(OUTPUT_FILE, folders_to_zip)
```
