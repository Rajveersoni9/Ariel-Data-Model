```python
import os
from pathlib import Path
from typing import Literal
import pandas as pd
```

```python
def remove_extra_data(data_path: Path, dataset:  Literal['train', 'test']) -> None:
    planet_ids = os.listdir(data_path / dataset)
    planet_ids = [int(planet_id) for planet_id in planet_ids if planet_id.isdigit()]
    
    star_info_path = data_path /  f'{dataset}_star_info.csv'
    if not star_info_path.exists():
        raise Exception(f"{star_info_path} does not exist.")
    
    star_info = pd.read_csv(star_info_path)
    star_info = star_info[star_info['planet_id'].isin(planet_ids)]
    assert len(star_info) == len(planet_ids), "Mismatch between planet IDs and star info entries."
    star_info.to_csv(star_info_path, index=False)
    
    if dataset == 'train':
        labels_path = data_path / 'train.csv'
        if not labels_path.exists():
            raise Exception(f"{labels_path} does not exist.")
        
        labels = pd.read_csv(labels_path)
        labels = labels[labels['planet_id'].isin(planet_ids)]
        assert len(labels) == len(planet_ids), "Mismatch between planet IDs and labels entries."
        labels.to_csv(labels_path, index=False)
```

```python
data_path = Path('../data/raw_subset')
remove_extra_data(data_path, 'train')
remove_extra_data(data_path, 'test')
```
